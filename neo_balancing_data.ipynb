```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
object = pd.read_csv("neo.csv")
```

```python
true_count = (object['hazardous'] == True).sum()
false_count = (object['hazardous'] == False).sum()

print("True count:", true_count)
print("False count:", false_count)
```

```python
from sklearn.utils import resample
import pandas as pd

haz_false = object[object['hazardous'] == False]
haz_true = object[object['hazardous'] == True]

haz_true_upsampled = resample(
    haz_true,
    replace=True,                 
    n_samples=len(haz_false),     
    random_state=42
)

balanced_data = pd.concat([haz_false, haz_true_upsampled])

balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

print(balanced_data['hazardous'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='hazardous', data=balanced_data)
plt.title("Balanced Dataset After Augmentation")
plt.show()
```
